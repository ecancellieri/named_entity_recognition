```python
import src.utils_ner as ner
import pandas as pd
```

# Load input strings

```python
inputdf = pd.read_csv('./input/Input_strings.csv')
display(inputdf)
```

# For each string:
## - assigns a category
## - checks if is "equivalent" to a previous one

```python
strings   = inputdf['String'].values
idcounter = {'Goods': 0, 'Entity': 0, 'Location': 0}
dfout     = pd.DataFrame(columns=['String', 'Category', 'GroupId'])

for string in strings:
    cat = ner.get_category(string)
    if cat in ['Goods', 'Entity', 'Location']:
        auxdf = dfout[dfout['Category'] == cat].copy()
        idcounter[cat], aux_groupid = ner.find_groupid_or_update_counter(string, auxdf, idcounter[cat])
        aux_groupid = cat+str(aux_groupid).replace(cat, '')
    else:
        aux_groupid = 'NaN'
    dfout.loc[len(dfout)] = [string, cat, aux_groupid]

print('Number of groups found:', idcounter)
display(dfout)
```
